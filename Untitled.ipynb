```python
import matplotlib.pyplot as plt
```

```python
x=["mon","tue","wed","thir","fri","satu","sun"]
y=[11,23,9,45,12,9,45]
plt.plot(x,y)
plt.title("sales according to day of week")
plt.xlabel("days of week")
plt.ylabel("sales of product")
plt.show()
```

```python
x1=["mon","tue","wed","thir","fri","satu","sun"]
y1=[11,23,9,45,12,9,45]
plt.plot(x1,y1,color="blue",linestyle="--",linewidth=2,marker='o',label="daywise sales")
plt.title("sales according to day of week")
plt.xlabel("days of week")
plt.ylabel("sales of product")
plt.grid(color='red',linestyle=':',linewidth=1)
# plt.Xlim(10,50)
plt.ylim(10,50)
plt.show()
```

```python
x2=[1,2,3,4,5]
y2=[34,56,78,90,23]
plt.plot(x2,y2,color="red",linewidth=2)
plt.grid(color="blue",linestyle=":",linewidth="1")
plt.xlabel("months")
plt.ylabel("tottal sales")
plt.title("this this sales data according to months")
plt.ylim(10,100)
plt.xticks([1,2,3,4,5],['m1','m2','m3','m4','m5'])
plt.show()
```

### bar chart

```python
product=['A','B','C','D','E','F']
sales=[1000,2500,900,1500,700,1100]
plt.bar(product,sales,color="orange",label="sales in 2025")
plt.xlabel("products name")
plt.ylabel("sales")
plt.title("product sales this year 2025")
plt.legend()
plt.show()
```

### pie chart

```python
regions=['North','South','East','West']
revenue=[300,200,1000,8000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','blue','lightgreen','coral'])
plt.title("revenue contribution by regions")
plt.show()
```

### Histogram

```python
score=[12,34,56,78,97,45,67,98,90,94,34,57,78,23,45,53,32,24]
plt.hist(score,bins=5,color="purple",edgecolor="black")
plt.xlabel("score range")
plt.ylabel("numbe of students")
plt.grid()
plt.show()
```

### scatter chart

```python
studied_hour=[2,5,6,7,8,3,9,4]
score=[20,50,60,70,80,30,100,40]
plt.scatter(studied_hour,score,color="green",marker="^",label="student data")  #o,s,^,*,+
plt.legend()
plt.xlabel("hour studeied")
plt.ylabel("marks")
plt.title("relationship between study time and marks")
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3,4,5,6],[30,40,50,60,70,80],color="orange",marker="^",label="class A")
plt.scatter([1,2,3,4,5,6],[50,60,70,80,90,100],color="blue",marker="^",label="class B")
plt.legend()
plt.xlabel("hour studeied")
plt.ylabel("marks")
plt.title("relationship between study time and marks")
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3,4,5,6],[30,40,50,60,70,80],color="orange",marker="^",label="class A")
plt.plot([1,2,3,4,5,6],[30,40,50,60,70,80])
plt.scatter([1,2,3,4,5,6],[50,60,70,80,90,100],color="blue",marker="^",label="class B")
plt.plot([1,2,3,4,5,6],[50,60,70,80,90,100])
plt.legend()
plt.xlabel("hour studeied")
plt.ylabel("marks")
plt.title("relationship between study time and marks")
plt.grid()
plt.show()
```

### subplot

```python
x=[1,2,3,4,5]
y=[234,56,678,900,322]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("bar chart")

plt.show()
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[23,56,76,10]

ax[0].plot(x,y)
ax[0].set_title('Line chart')

ax[1].bar(x,y)
ax[1].set_title('Bar chart')

fig.suptitle("comparision of line and bar chart")
plt.show()
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[23,56,76,10]

ax[0].plot(x,y)
ax[0].set_title('Line chart')

ax[1].bar(x,y)
ax[1].set_title('Bar chart')

fig.suptitle("comparision of line and bar chart")
plt.tight_layout()
plt.savefig("comparision.png",dpi=300,bbox_inches="tight")
plt.show()
```
